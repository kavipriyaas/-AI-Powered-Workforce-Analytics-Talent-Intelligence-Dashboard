create the e-commerce transaction dataset 10 rows and 10 columns using sql

```sql
-- Create the E-commerce Transactions table
CREATE TABLE e_commerce_transactions (
    transaction_id VARCHAR(50) PRIMARY KEY,
    customer_id VARCHAR(50) NOT NULL,
    product_id VARCHAR(50) NOT NULL,
    quantity INT NOT NULL,
    price DECIMAL(10, 2) NOT NULL,
    total_amount DECIMAL(10, 2) GENERATED ALWAYS AS (quantity * price) STORED,
    transaction_date DATE NOT NULL,
    payment_method VARCHAR(50) NOT NULL,
    shipping_city VARCHAR(50) NOT NULL,
    product_category VARCHAR(50) NOT NULL
);

-- Insert 10 rows of sample data
INSERT INTO e_commerce_transactions (
    transaction_id, customer_id, product_id, quantity, price, transaction_date, payment_method, shipping_city, product_category
) VALUES
('T001', 'C101', 'P001', 2, 25.50, '2023-01-10', 'Credit Card', 'New York', 'Electronics'),
('T002', 'C102', 'P002', 1, 120.00, '2023-01-10', 'PayPal', 'Los Angeles', 'Books'),
('T003', 'C103', 'P003', 3, 15.00, '2023-01-11', 'Debit Card', 'Chicago', 'Apparel'),
('T004', 'C101', 'P004', 1, 300.00, '2023-01-11', 'Credit Card', 'New York', 'Electronics'),
('T005', 'C104', 'P005', 5, 5.99, '2023-01-12', 'Bank Transfer', 'Houston', 'Food & Beverage'),
('T006', 'C102', 'P006', 2, 80.00, '2023-01-12', 'PayPal', 'Los Angeles', 'Home Goods'),
('T007', 'C105', 'P001', 1, 25.50, '2023-01-13', 'Credit Card', 'Phoenix', 'Electronics'),
('T008', 'C103', 'P007', 4, 10.00, '2023-01-13', 'Debit Card', 'Chicago', 'Apparel'),
('T009', 'C106', 'P002', 1, 120.00, '2023-01-14', 'Credit Card', 'New York', 'Books'),
('T010', 'C104', 'P008', 2, 45.00, '2023-01-14', 'Bank Transfer', 'Houston', 'Home Goods');

-- Optional: Select all data to verify
-- SELECT * FROM e_commerce_transactions;
```
