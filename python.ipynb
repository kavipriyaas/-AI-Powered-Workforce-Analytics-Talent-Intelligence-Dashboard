create the e-commerce transaction dataset 10 rows and 10 columns

```python
import pandas as pd
import numpy as np

def create_ecommerce_dataset(num_rows, num_cols):
    np.random.seed(42) # for reproducibility
    product_categories = ['Electronics', 'Books', 'Clothing', 'Home Goods', 'Beauty']
    payment_methods = ['Credit Card', 'PayPal', 'Bank Transfer', 'Cash on Delivery']
    shipping_status = ['Pending', 'Shipped', 'Delivered', 'Cancelled']

    data = {
        'TransactionID': range(1000, 1000 + num_rows),
        'UserID': np.random.randint(1, 100, num_rows),
        'ProductID': np.random.randint(100, 500, num_rows),
        'Quantity': np.random.randint(1, 5, num_rows),
        'Price': np.round(np.random.uniform(5.0, 500.0, num_rows), 2),
        'Category': np.random.choice(product_categories, num_rows),
        'PaymentMethod': np.random.choice(payment_methods, num_rows),
        'ShippingStatus': np.random.choice(shipping_status, num_rows),
        'TransactionDate': pd.to_datetime(pd.date_range(start='2023-01-01', periods=num_rows, freq='D')),
        'TotalAmount': np.round(np.random.uniform(10.0, 2000.0, num_rows), 2)
    }

    # Select first num_cols columns if more columns are generated than requested
    df = pd.DataFrame(data).iloc[:, :num_cols]
    return df

# Create the dataset with 10 rows and 10 columns
ecommerce_df = create_ecommerce_dataset(10, 10)
print(ecommerce_df)
```

```python
import pandas as pd

# Calculate TotalItemPrice
ecommerce_df['TotalItemPrice'] = ecommerce_df['Quantity'] * ecommerce_df['Price']

# Extract Day of Week from TransactionDate
ecommerce_df['DayOfWeek'] = ecommerce_df['TransactionDate'].dt.day_name()

# Extract Month from TransactionDate
ecommerce_df['Month'] = ecommerce_df['TransactionDate'].dt.month_name()

print(ecommerce_df.head())
```
